```python
import speech_recognition as sr
import pandas as pd
import re

# Initialize recognizer and load Excel
recognizer = sr.Recognizer()
excel_file = "Book2.xlsx"
df = pd.read_excel(excel_file)

def process_audio(audio_path):
    """Process audio input and return split values"""
    with sr.AudioFile(audio_path) as source:
        audio = recognizer.record(source)

    try:
        raw_text = recognizer.recognize_google(audio)
        print(f"Raw recognition: {raw_text}")  # Debug output

        # Split entries on "next" (case-insensitive) and clean
        entries = [entry.strip() for entry in re.split(r'\bnext\b', raw_text, flags=re.IGNORECASE) if entry.strip()]

        processed_entries = []
        for entry in entries:
            # Clean and validate numerical patterns
            cleaned = re.sub(r'[,\s]', '', entry)
            if re.match(r'^(\d+\.?\d*|(\d+\.?\d*-\d+\.?\d*)|(\d+°)|(\d+-\d+))$', cleaned):
                processed_entries.append(cleaned)
            else:
                processed_entries.append(entry)  # Preserve non-numerical entries

        return processed_entries

    except Exception as e:
        print(f"Processing error: {e}")
        return []

def excel_writer(values):
    """Write multiple values to Excel with format preservation"""
    if not values:
        print("No valid input detected")
        return

    for value in values:
        # Convert numerical strings to float if simple format
        if re.match(r'^\d+\.?\d*$', value):
            write_value = float(value)
        else:
            write_value = str(value)

        # Append to Excel
        df.loc[len(df)] = [write_value]

    df.to_excel(excel_file, index=False)
    print(f"Successfully wrote {len(values)} entries")

# Usage example
audio_path = "/content/audio.wav"
processed_values = process_audio(audio_path)
excel_writer(processed_values)
```

```python
!pip install openai-whisper pandas openpyxl word2number
import ipywidgets as widgets
from IPython.display import display, Javascript, clear_output
import pandas as pd
import whisper
import os
from word2number import w2n
import re

# Use larger model for better number recognition
model = whisper.load_model("base")  # Changed from "tiny" to "base"

excel_file = "/content/live_entries.xlsx"

# Create Excel file if it doesn't exist
if not os.path.exists(excel_file):
    pd.DataFrame(columns=['Entry']).to_excel(excel_file, index=False)

def convert_numbers(text):
    """Convert spoken numbers to digits while preserving text"""
    # Handle decimal patterns: "point" -> ".", "one point five" -> 1.5
    text = re.sub(r'(\d+)\s+point\s+(\d+)', r'\1.\2', text)
    text = re.sub(r'point\s+(\d+)', r'.\1', text)

    # Convert number words to digits using word2number
    words = text.split()
    converted = []
    current_number = []

    for word in words:
        # Try to parse number words/digits
        if word.replace(',', '').replace('.', '').isdigit():
            converted.append(word)
        else:
            try:
                # Handle ordinal numbers ("first" -> 1)
                if word.endswith(('st', 'nd', 'rd', 'th')):
                    num_word = word[:-2]
                    num = str(w2n.word_to_num(num_word))
                else:
                    num = str(w2n.word_to_num(word))

                current_number.append(num)
            except:
                # Flush accumulated numbers and add text
                if current_number:
                    converted.append(''.join(current_number))
                    current_number = []
                converted.append(word)

    # Add any remaining numbers
    if current_number:
        converted.append(''.join(current_number))

    # Handle number ranges: "25 to 30" -> "25-30"
    converted_text = ' '.join(converted)
    converted_text = re.sub(r'(\d+)\s+(to|dash|through)\s+(\d+)', r'\1-\3', converted_text)

    return converted_text

# JavaScript recorder code remains the same
recorder_js = """
const sleep = ms => new Promise(r => setTimeout(r, ms));
const record = (duration) => {
    return new Promise(async resolve => {
        const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
        const recorder = new MediaRecorder(stream);
        const chunks = [];
        recorder.ondataavailable = e => chunks.push(e.data);
        recorder.start();
        await sleep(duration * 1000);
        recorder.stop();
        stream.getTracks().forEach(track => track.stop());
        recorder.onstop = () => resolve(new Blob(chunks));
    });
};
"""

button = widgets.Button(description="Start Recording (5s)")
output = widgets.Output()

def on_button_click(b):
    clear_output()
    print("Recording...")
    display(Javascript(recorder_js + """
    (async function() {
        const blob = await record(5);
        const reader = new FileReader();
        reader.readAsDataURL(blob);
        reader.onloadend = () => {
            const audio_data = reader.result.split(',')[1];
            google.colab.kernel.invokeFunction('process_audio', [audio_data], {});
        };
    })();
    """))

button.on_click(on_button_click)
display(button, output)

def process_audio(audio_data):
    import base64

    audio_bytes = base64.b64decode(audio_data)
    with open("temp.webm", "wb") as f:
        f.write(audio_bytes)

    result = model.transcribe("temp.webm", fp16=False)
    raw_text = result["text"].strip()

    # Convert numbers while preserving text
    processed_text = convert_numbers(raw_text)

    # Split entries and clean
    entries = [entry.strip() for entry in processed_text.split('next') if entry.strip()]

    if entries:
        df = pd.read_excel(excel_file)
        for entry in entries:
            # Preserve numeric types where possible
            try:
                entry = float(entry) if '.' in entry else int(entry)
            except ValueError:
                pass
            df.loc[len(df)] = [entry]

        df.to_excel(excel_file, index=False)
        print(f"Saved entries: {entries}")
    else:
        print("No valid entries found")

from google.colab import output
output.register_callback('process_audio', process_audio)
```
